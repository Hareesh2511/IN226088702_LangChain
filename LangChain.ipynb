```python
prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)
```

```python
chain = prompt | llm | StrOutputParser()
```

```python
def calculator_tool(query):
    return "This is a simple calculator tool (demo output -->  4)"

tools = {
    "calculator": calculator_tool
}
```

```python
def agent(user_input):

    # If user asks calculation → use tool
    if "calculate" in user_input:
        return tools["calculator"](user_input)

    # Otherwise → use LLM chain
    return chain.invoke({"topic": user_input})
```

```python
user_input = input("Enter your query: ")

output = agent(user_input)

print("\nFinal Output:")
print(output)
```
